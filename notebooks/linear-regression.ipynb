```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
X = np.array([1,2,3])
Y = np.array([1, 2.5, 3.5])
```

```python
def calcCost(hx,y,m, theta1):
    factor = 1/(2*len(hx))
    return factor * np.sum(pow(hx - y,2))
```

```python
def hypothesis(X,theta1):
    return X * theta1
```

```python
def plotData(x, y, h):
    ax = plt.subplots()
    plt.plot(X, h)
    plt.scatter(x, y, c='red')
    plt.grid(color='gray', linestyle='--', linewidth=.6, axis='both', which='both', alpha=.4)
    plt.show()
```

```python
step=0.05
theta = .75
H = hypothesis(X , theta)
cost = calcCost(H,Y,len(X),theta)
finalTheta = theta
for i in range(0, 100):
    theta += step
    hx = hypothesis(X , theta)
    c = calcCost(hx,Y,len(X),theta)
    if c < cost:
        cost = c
        H = hx
        finalTheta=theta
print("Theta : {}, Cost : {}".format(finalTheta,cost))
plotData(X,Y,H)
```
