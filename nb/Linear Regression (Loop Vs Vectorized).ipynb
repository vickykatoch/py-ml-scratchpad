```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from IPython.core.pylabtools import figsize
pd.options.display.float_format = '{:,.6f}'.format
np.set_printoptions(precision=15,suppress=True)
import helpers as helper
```

```python
df = pd.read_csv('../data/housing.data', header=None, names=helper.getHousingDataColumnNames(), sep='\s+')
(df, X, y, m, n) = helper.getInitialData(df)
theta = np.ones((n,1))
h = helper.hypothesis_l(X,theta)
alpha = 0.0005
iterations = 1000
```

### 1. Linear Regression using Gradient Descent (Loop)

```python
%%time
cost = np.zeros((iterations,2))
initial_theta=theta
for i in range(iterations):
    hyp = helper.hypothesis_l(X,initial_theta)
    cost = helper.cost_l(hyp,y)
    initial_theta = helper.gradient_descent_l(X,h,y, alpha,initial_theta)
# initial_theta
```

### 2. Linear Regression using Gradient Descent (Vectorized)

```python
%%time
cost = np.zeros((iterations,2))
initial_theta=theta
for i in range(iterations):
    hyp = helper.hypothesis_v(X,initial_theta)
    cost = helper.cost_v(hyp,y)
    initial_theta = helper.gradient_descent_v(X,h,y, alpha,initial_theta)
# initial_theta
```
